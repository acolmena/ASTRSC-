### Adding More Words to Conflict Array by Repeating MediaCloud Analysis
We decided to do one last MediaCloud analysis to see if there are any more words we can add to the ConflictViolenceArray. From our MediaCloud search, we obtained several .csv files with words commonly found in relevant articles. We will copy and paste these words to an array, delete any repeated entries (among the new and old words), and manually select new unique words we can add to the array.

```python
# import relevant libraries
import numpy as np
import pandas as pd
```

```python
# add current words to array
curr_array = [
    "violence",
    "conflict",
    "war",
    "atrocious",
    "atrocity",
    "crime",
    "violent",
    "weapon",
    "insurgency",
    "insurgent",
    "insurgence",
    "chaos",
    "chaotic",
    "arm",
    "terror",
    "flee",
    "militia",
    "outbreak",
    "crisis",
    "brutal",
    "collapse",
    "tension",
    "rebel",
    "genocide",
    "tension",
    "clash",
    "dead",
    "eruption",
    "criminal",
    "terrorism",
    "extremist",
    "abuse",
    "escalation",
    "escalate",
    "rape",
    "sanction",
    "corruption",
    "corrupt",
    "radical",
    "ban",
    "bomb",
    "blood",
    "cease-fire",
    "coup",
    "assault",
    "kidnap",
    "battle",
  ]

all_new_wrds = []
all_new_roots = []


filenames = ["conflict-africa-violence-sampled-1000-ngrams-1-20221101182330.csv", "conflict-and-car-and-violence-sampled-1000-ngrams-1-20201019183803.csv", "conflict-and-congo-and-violenc-sampled-1000-ngrams-1-20201019183458.csv", "conflict-and-darfur-and-vio-sampled-1000-ngrams-1-20201019182856.csv", "conflict-and-kenya-and-viol-sampled-1000-ngrams-1-20201019183144.csv", "war-violence-africa-sampled-1000-ngrams-1-20221101220647.csv"]

# 1. Read .csv file using numpy
for filename in filenames:
    file_wrds_roots = np.loadtxt(filename, delimiter=",", dtype=str, skiprows=1, usecols=(0,1))

    # create arrays
    new_array = np.array(file_wrds_roots)
    new_wrds = np.array(new_array[:,0])
    new_roots = np.array(new_array[:,1])
    
    # remove double quotes from entries
    new_wrds = np.char.replace(new_wrds, '"', '') 
    new_roots = np.char.replace(new_roots, '"', '') 
    
    # 2. Combine new arrays with each other and with cur array 
    all_new_wrds = np.concatenate([all_new_wrds, new_wrds], axis=None)
    all_new_roots = np.concatenate([all_new_roots, new_roots])
    
# print(all_new_wrds, len(all_new_wrds), len(array))

# 3. Filter out all repeated words in combo array 
unique_new_wrds = np.unique(all_new_wrds)
unique_new_roots = np.unique(all_new_roots)
# print(len(unique_new_wrds), len(unique_new_roots))

not_in_curr_unique_wrds = []
for uniq_wrd in unique_new_wrds:
    if uniq_wrd not in curr_array and uniq_wrd.isalpha():
        not_in_curr_unique_wrds.append(uniq_wrd)

# not_in_curr_unique_wrds = np.delete(unique_new_wrds, np.where(unique_new_wrds in curr_array))
# not_in_curr_unique_wrds = unique_new_wrds[unique_new_wrds not in curr_array]

print(len(not_in_curr_unique_wrds), not_in_curr_unique_wrds)

# 4. Convert array to dataframe and export as .csv file
DF = pd.DataFrame(not_in_curr_unique_wrds)
DF.to_csv("new_mediacloud_wrds.csv")


# 5. Manually pick out new words that are relevant to ConflictViolenceArray
```
