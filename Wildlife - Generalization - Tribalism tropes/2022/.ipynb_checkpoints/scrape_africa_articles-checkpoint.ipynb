```python
# necessary packages
import re
from bs4 import BeautifulSoup as bs
import requests
```

## Scrape links to articles about Africa from CNN, BBC, NBC, Fox News, Al Jazeera
We will now make an array of all the links of the news outlets' website pages containing all recent articles written about Africa. 

```python
urls = ["https://www.cnn.com/africa", "https://www.bbc.com/news/world/africa", "https://www.nbcnews.com/search/?q=africa",  "https://www.foxnews.com/category/world/world-regions/africa", "https://www.aljazeera.com/africa/"]

news_outlets = ["cnn", "bbc", "nbc", "fox", "aljazeera"]
html_soups = {}

for i in range(5):
    # send request
    res = requests.get(urls[i])

    txt = res.text
    status = res.status_code

    # convert to a beautiful soup object
    soup = bs(res.content, 'html.parser')
    
    # store soup in dict
    html_soups[news_outlets[i]] = soup
```
